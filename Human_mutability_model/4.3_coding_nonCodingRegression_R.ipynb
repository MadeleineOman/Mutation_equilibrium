# **perfomring the regression comparing the triplet mutability in coding vs noncoding**

```r
library(ggplot2)
```

```r
data <- read.csv("wrangled_codingNoncodingMutability_forRgression_robfile.csv")
```

```r
model <- lm(coding_mut~noncoding_mut, data = data)
```

```r
summary(model)
```

```r
ggplot(data, aes(x=coding_mut,y=noncoding_mut))+
    labs( y="Noncoding model mutability" , x="Coding model mutability")+
    geom_point()
# ggsave("plots/codingVsNoncodingMut_2021_09_17_nointrons.png")
```
