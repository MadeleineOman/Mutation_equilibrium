```python
Count_dataDict = {}
sim_name_afterSim = "Triplet_mut_counts_Trial"
about_sim = ""
file_path_afterSim = ""

nTrials = 1000

#DifDNA800kgen trial 1-5
for index_tmp in tqdm(range(1,int(nTrials)+1)):

    file_name = sim_name_afterSim+str(index_tmp)+".txt"

    file_object_dict = json.loads(open(file_path_afterSim+file_name).readlines()[0])
    
    Count_dataDict["Trial"+str(index_tmp)] = file_object_dict


```

```python
triplets_averaged_counts = []
triplet_sd_counts = []
for triplet in triplets: 
    tmp_triplet_count_list = []
    for trial_number in range(1, nTrials+1): 
        trial_name = "Trial"+str(trial_number)
        
        tmp_triplet_count_list.append(Count_dataDict[trial_name][triplet])
    triplets_averaged_counts.append(np.mean(tmp_triplet_count_list))
    triplet_sd_counts.append(np.std(tmp_triplet_count_list)/np.sqrt(nTrials))
```
