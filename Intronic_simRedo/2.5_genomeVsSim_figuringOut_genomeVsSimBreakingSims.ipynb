# ** data visualization: mutation count relfect triplet freq and vice versa ** 

```python
import matplotlib.pyplot as plt
from datetime import datetime
import numpy as np 
import json
```

```python
model0 = json.load(open("../Human_mutability_model/Model_2020_12_02_genomeWide.txt"))
```

```python
mut_dict = json.load(open("../Human_mutability_model/MutDict_2020_12_10.txt"))
```

```python
triplet_genome_dict = json.load(open("../HumanTripletCounts/GenomeTripletCounts_2020_12_02_genomeWide.txt"))
```

```python
triplets = []
muts_count = []
mutabilities = []
triplet_genomeCount = []
for key in mut_dict.keys(): 
    triplets.append(key)
    muts_count.append(mut_dict[key][0])
    triplet_genomeCount.append(int(triplet_genome_dict[key]))
    mutabilities.append(model0[key][0])
```

```python
plt.style.use("seaborn-ticks")
```

```python
plt.scatter(muts_count, triplet_genomeCount )
plt.xlabel("Mutation dataset triplet count")
plt.ylabel("Genome triplet count")


plt.show()
```

```python
plt.hist(muts_count)
plt.show()
```

```python
plt.hist(triplet_genomeCount)
plt.show()
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.bar([i for i in range(len(muts_count))],muts_count, tick_label = triplets)
# plt.xlabel("Mutation dataset triplet count")
# plt.ylabel("Genome triplet count")


plt.show()
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.bar([i for i in range(len(muts_count))],triplet_genomeCount, tick_label = triplets)
# plt.xlabel("Mutation dataset triplet count")
# plt.ylabel("Genome triplet count")


plt.show()
```
