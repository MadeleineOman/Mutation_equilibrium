# **I only need the 0.1x coverage counts plot for the SMBE2021 poster**

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from datetime import datetime
```

```python
df = pd.read_csv("data/Data_dataframes/final_plotting_01_2x_coverage_2021_04_21.csv")
```

```python
df.head()
```

```python
sns.set_context("paper", font_scale=2.2)
plt.figure(figsize=(10,9))

plot = sns.scatterplot(x="mut_log10", y="countsInto_01_mean", data=df, color = "grey",s=50,edgecolor = "grey", label = "created")
plot = sns.scatterplot(x="mut_log10", y="countsChosen_01_mean", data=df, color = "k",s=50,edgecolor = "grey",label="chosen")

plot.set(xlabel = "Mutability (log10)", ylabel = "Triplet counts")


# plt.show()
timestamp = datetime.now().strftime("%Y/%m/%d").replace("/", "_").replace(":", "_")
plt.savefig("plots/Scatter_{t}_{l}Mbp_{p}propmuts_accurateINTO_genomeModel_log_PosterSMBE.png".format(p = 0.1, l = 0.1, t = timestamp))
```
